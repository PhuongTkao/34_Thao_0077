3. Thống kê mô tả:

-Các đại lượng thống kê cơ bản:

+  Trung bình (Mean): Trung bình số lượng huy chương mà Việt Nam đạt được mỗi năm.

+  Trung vị (Median): Giá trị chính giữa của số lượng huy chương trong giai đoạn phân tích

+  Phương sai (Variance) và Độ lệch chuẩn (Standard Deviation): Đo lường mức độ biến động của số huy   chương qua các năm

-Phân tích phân phối dữ liệu: Để đánh giá đặc điểm phân phối của số huy chương đạt được qua các năm:

+ Sử dụng ước lượng tỷ lệ: Xác định xác suất đạt được từng loại huy chương trong các kỳ thi, giúp dự đoán xu hướng thành tích của Việt Nam.

+ Phân tích mật độ xác suất (Probability Density Estimation): Kiểm tra xem số lượng huy chương có tập trung vào một khoảng nhất định hay không


-Sử dụng ước lượng mô hình:

+ Hồi quy tuyến tính đơn biến: Dự đoán số lượng huy chương dựa trên một yếu tố như số lượng thí sinh tham gia.

+ Hồi quy tuyến tính đa biến: Xem xét ảnh hưởng của nhiều yếu tố như thời gian chuẩn bị, mức độ khó của đề thi đến kết quả huy chương.

+ Mô hình ARDL(Autoregressive Distributed Lag Model): Đánh giá mối quan hệ dài hạn và ngắn hạn giữa số huy chương và các yếu tố như chính sách đào tạo.

+ Mô hình chuỗi thời gian: (Time Series Models - ARIMA, SARIMA, VAR): Dự báo số huy chương tương lai dựa trên dữ liệu lịch sử.

+ Mô hình học máy (Machine Learning - Random Forest, XGBoost, Neural Networks): Xác định mối quan hệ phi tuyến tính và dự báo số huy chương với độ chính xác cao.


-Thực hiện kiểm định thống kê:

+ Kiểm định Anova: Xác định xem có sự khác biệt đáng kể giữa số lượng huy chương qua các giai đoạn phát triển khác nhau hay không (ví dụ: trước và sau một chính sách đào tạo mới).

+ Kiểm định Shapiro-Wilk (Shapiro-Wilk Test): Kiểm tra xem số lượng huy chương có phân phối chuẩn hay không, nếu không thì có thể sử dụng phương pháp phân tích phi tham số thay thế.

+ Kiểm định Chi-square: Kiểm tra xem có mối liên hệ giữa số lượng huy chương đạt được và các yếu tố như năm thi, số lượng thí sinh tham gia, hoặc quốc gia so sánh, từ đó đánh giá yếu tố nào có ảnh hưởng mạnh nhất.

Trong dữ liệu, các thí sinh đạt thành tích Bằng khen bị thiếu thông tin về Điểm số và Hạng. Để đảm bảo tính nhất quán và phản ánh đúng thứ bậc thành tích, áp dụng cách điền dựa trên điểm và hạng của Huy chương Đồng – giải thưởng liền kề cao hơn Bằng khen:

- Điểm số:
Tôi xác định điểm số thấp nhất của nhóm thí sinh đạt Huy chương Đồng từng năm.
Sau đó, tôi điền điểm số cho Bằng khen bằng cách lấy 80% giá trị đó (tức thấp hơn 20%), đảm bảo Bằng khen luôn có điểm số thấp hơn huy chương đồng.
- Hạng:
Tôi xác định hạng cao thấp nhất của Huy chương Đồng.
Điền hạng cho Bằng khen bằng cách cộng thêm một khoảng hạng (ví dụ: +15 hạng) so với hạng Huy chương Đồng.
Điều này hợp lý vì thí sinh Bằng khen xếp sau Huy chương Đồng, thể hiện đúng phân tầng kết quả.

```r
for (year in c(2002, 2009)) {

  # ---- 1. XỬ LÝ ĐIỂM SỐ ----
  dong_diem <- data$`Điểm số`[data$`IOI (năm)` == year & data$`Giải thưởng` == "HCĐ"]

  if (length(dong_diem[!is.na(dong_diem)]) > 0) {
    min_diem_dong <- min(dong_diem, na.rm = TRUE)

    # Vị trí Bằng khen bị NA điểm
    idx_diem <- which(data$`IOI (năm)` == year & data$`Giải thưởng` == "Bằng khen" & is.na(data$`Điểm số`))

    # Gán điểm = 80% điểm đồng nhỏ nhất
    data$`Điểm số`[idx_diem] <- min_diem_dong * 0.8
  }

  # ---- 2. XỬ LÝ HẠNG ----
  dong_hang <- data$`Hạng`[ data$`IOI (năm)` == year & data$`Giải thưởng` == "HCĐ"]

  if (length(dong_hang[!is.na(dong_hang)]) > 0) {
    min_hang_dong <- min(dong_hang, na.rm = TRUE)

    # Vị trí Bằng khen bị NA hạng
    idx_hang <- which( data$`IOI (năm)` == year & data$`Giải thưởng` == "Bằng khen" & is.na(data$`Hạng`))

    # Gán hạng = hạng đồng nhỏ nhất - 15
    data$`Hạng`[idx_hang] <- min_hang_dong - 15
  }
}

# Chỉ hiển thị lại các năm 2002 và 2009
df_result <- data[ data$`IOI (năm)` %in% c(2002, 2009), ]

df_result

```

**c.Trung vị (Median)** là giá trị trung tâm của một tập dữ liệu khi sắp xếp theo thứ tự tăng dần. Trong phân tích dữ liệu Olympic Tin học Quốc tế, trung vị có thể dùng để xác định số huy chương điển hình mà Việt Nam đạt được mỗi năm.

Tính tổng số huy chương mỗi năm bằng cách đếm số lần xuất hiện của từng năm.

aggregate(Giai_thuong ~ Nam, data, length):

**d.Phương sai (Variance)** đo lường mức độ phân tán của số huy chương quanh giá trị trung bình. Giá trị phương sai càng lớn, số huy chương giữa các năm càng biến động.

**Độ lệch chuẩn (Standard Deviation)** là căn bậc hai của phương sai, cho biết mức độ dao động trung bình của số huy chương so với giá trị trung bình. Giá trị càng lớn, sự biến động càng cao.

var(): Tính phương sai (độ phân tán của dữ liệu so với giá trị trung bình).

sd(): Tính độ lệch chuẩn (mức độ chênh lệch trung bình của dữ liệu so với trung bình cộng).

**c.  So sánh tỷ lệ huy chương giữa các khu vực địa lý**

**Mục tiêu**

-So sánh thành tích của Việt Nam khi thi đấu tại các khu vực để xác định liệu có lợi thế hơn khi thi trong khu vực đó hay không.

-Đánh giá mức độ cạnh tranh giữa các khu vực khác nhau.


```r
# Nạp thư viện cần thiết
library(dplyr)

# Định nghĩa danh sách các khu vực
nhom_khu_vuc <- list(
  "Đông Nam Á" = c("Vietnam", "Thailand", "Singapore", "Malaysia", "Indonesia", "Philippines"),
  "Châu Âu" = c("Russia", "Germany", "United Kingdom", "France", "Italy"),
  "Bắc Mỹ" = c("United States", "Canada"),
  "Châu Á" = c("China", "Japan", "South Korea", "India"),
  "Khác" = c("Australia", "Brazil", "Mexico") # Các nước không thuộc nhóm trên
)

# Thêm cột Khu_vuc vào data dựa trên danh sách nhóm khu vực
data$Khu_vuc <- sapply(data$Quoc_gia_to_chuc, function(x) {
  for (kv in names(nhom_khu_vuc)) {
    if (x %in% nhom_khu_vuc[[kv]]) return(kv)
  }
  return("Khác") # Nếu không khớp với nhóm nào thì xếp vào "Khác"
})

# Đếm số lượng huy chương theo khu vực
huy_chuong_theo_khu_vuc <- table(data$Khu_vuc, data$Giai_thuong)

# Tính tỷ lệ huy chương theo khu vực
ty_le_huy_chuong_khu_vuc <- prop.table(huy_chuong_theo_khu_vuc, margin = 1) * 100

# Chuyển thành data frame
df_tl_hc_khu_vuc <- as.data.frame(as.table(ty_le_huy_chuong_khu_vuc))

# Đổi tên cột
colnames(df_tl_hc_khu_vuc) <- c("Khu_vuc", "Giai_thuong", "Ty_le")

# Lọc bỏ các hàng có tỷ lệ 0%
df_tl_hc_khu_vuc <- df_tl_hc_khu_vuc %>% filter(Ty_le > 0)

# Chuẩn hóa lại tỷ lệ theo từng khu vực
df_tl_hc_khu_vuc <- df_tl_hc_khu_vuc %>%
  group_by(Khu_vuc) %>%
  mutate(Ty_le = Ty_le / sum(Ty_le) * 100) %>%
  ungroup()

# In kết quả
print("Tỷ lệ huy chương theo khu vực địa lý:")
print(df_tl_hc_khu_vuc)
```

**Nhận xét:**

-Nếu tỷ lệ huy chương Vàng cao hơn ở một số khu vực nhất định → Có thể môi trường thi đấu tại khu vực đó thuận lợi hơn cho thí sinh Việt Nam

-Nếu tỷ lệ huy chương thấp hơn ở các khu vực khác → Có thể mức độ cạnh tranh cao hơn, phong cách ra đề khác biệt hoặc các yếu tố bên ngoài như khoảng cách địa lý và điều kiện sinh hoạt ảnh hưởng đến phong độ.

**Dự đoán xu hướng:**

-Nếu Việt Nam tiếp tục có tỷ lệ huy chương cao ở một khu vực nhất định → Có thể tập trung tối ưu chiến lược để duy trì lợi thế, đồng thời tận dụng cơ hội để cải thiện thành tích ở các khu vực khác.

-Nếu tỷ lệ huy chương tăng lên ở các khu vực trước đây có thành tích thấp → Chứng tỏ thí sinh Việt Nam đang dần thích nghi với môi trường thi đấu đa dạng, mở rộng cơ hội đạt thành tích cao hơn trên phạm vi toàn cầu.

c. Kiểm định phân phối chuẩn

 Trước khi thực hiện kiểm định ANOVA, chúng tôi kiểm tra phân phối chuẩn của điểm thi cho mỗi năm bằng kiểm định Shapiro-Wilk.

---


**a. Nhóm theo thời gian: So sánh số lượng huy chương giữa các giai đoạn (ví dụ: 1989-2000, 2001-2010, 2011-2024).**

**Mục tiêu**

-Xác định xem số lượng huy chương có khác biệt đáng kể giữa các giai đoạn hay không.

-Nếu có sự khác biệt, có thể xem xét nguyên nhân như chính sách đào tạo, số lượng thí sinh tham gia, v.v.

**Cách thực hiện**

-Nhóm dữ liệu theo Giai đoạn (1989-2000, 2001-2010, 2011-2024).

-Tính tổng số huy chương trong từng giai đoạn.

-Kiểm tra phân phối chuẩn của dữ liệu bằng Shapiro-Wilk Test.

-Nếu dữ liệu có phân phối chuẩn → Dùng ANOVA để kiểm tra sự khác biệt.

-Nếu dữ liệu không có phân phối chuẩn → Dùng Kruskal-Wallis Test (kiểm định phi tham số).



**Bước 1:** Nhóm dữ liệu theo giai đoạn

-Trước tiên, cần gán giai đoạn cho từng năm:




```r
library(dplyr)

# Gán giai đoạn cho từng năm
data <- data %>%
  mutate(Giai_doan = case_when(
    Nam >= 1989 & Nam <= 2000 ~ "1989-2000",
    Nam >= 2001 & Nam <= 2010 ~ "2001-2010",
    Nam >= 2011 & Nam <= 2024 ~ "2011-2024"
  ))

# Đếm số lượng huy chương theo giai đoạn
huy_chuong_theo_giai_doan <- data %>%
  group_by(Giai_doan) %>%
  summarise(So_huy_chuong = n(), .groups = "drop")

print(huy_chuong_theo_giai_doan)
```

**Bước 2:** Kiểm tra phân phối chuẩn với Shapiro-Wilk Test

```r

shapiro.test(huy_chuong_theo_giai_doan$So_huy_chuong)
```

**Kết quả**

-Nếu p-value > 0.05 → Dữ liệu có phân phối chuẩn.

-Nếu p-value < 0.05 → Dữ liệu không có phân phối chuẩn.

**Bước 3:** Kiểm định ANOVA (nếu dữ liệu có phân phối chuẩn)



```r
anova_result <- aov(Nam ~ Giai_doan, data = data)
summary(anova_result)
```

-Nếu p-value < 0.05 → Có sự khác biệt đáng kể giữa các giai đoạn.



**Bước 4:** Kiểm định Kruskal-Wallis (nếu dữ liệu không có phân phối chuẩn)

-Nếu dữ liệu không có phân phối chuẩn, ta thay thế ANOVA bằng Kruskal-Wallis Test:





```r
# Bước 4: Nếu dữ liệu không có phân phối chuẩn, sử dụng Kruskal-Wallis
if (shapiro_result$p.value < 0.05) {
  print("Dữ liệu không có phân phối chuẩn, tiến hành kiểm định Kruskal-Wallis:")
  kruskal_result <- kruskal.test(So_huy_chuong ~ Giai_doan, data = huy_chuong_theo_nam)
  print(kruskal_result)
} else {
  print("Dữ liệu có phân phối chuẩn, kết quả ANOVA được sử dụng.")
}
```

-Nếu p-value < 0.05 → Có sự khác biệt giữa các giai đoạn.



**Kết luận:**

-Nếu có sự khác biệt, cần phân tích nguyên nhân như chính sách đào tạo, số lượng thí sinh, v.v.

-Nếu không có sự khác biệt, có thể kết luận số lượng huy chương không bị ảnh hưởng đáng kể bởi các giai đoạn.

Tuy nhiên, mức độ giảm này không đáng kể, và khoảng tin cậy khá rộng, cho thấy sự biến động lớn trong dữ liệu.
```r
# Load thư viện
library(dplyr)
library(ggplot2)
library(readr)
url <- "https://raw.githubusercontent.com/hoainam5/DO_AN_1_NHOM_1_3_DHKL17A3HN/refs/heads/master/data_IOI.csv"

# Đọc file CSV
data <- read_csv(url)

# Vẽ biểu đồ hộp so sánh điểm số (Diem_so) theo từng năm (Nam)
ggplot(data, aes(x = factor(Nam), y = Diem_so)) +
  geom_boxplot(aes(fill = factor(Nam)), outlier.colour = "red", outlier.shape = 16) +
  labs(
    title = "Biểu đồ hộp: Phân bố điểm số của học sinh Việt Nam tại IOI (1989 - 2024)",
    x = "Năm",
    y = "Điểm số"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1),  # Xoay nhãn năm để dễ đọc
    legend.position = "none"                             # Ẩn phần chú giải màu
  )
```

**Nhận xét và phân tích:** Qua biểu đồ, ta nhận thấy rằng trong giai đoạn 1989–2000 có nhiều năm điểm số dao động mạnh với một số năm có trung vị rất cao, trong khi giai đoạn 2001–2010 có xu hướng giảm nhẹ điểm số nhưng độ phân tán tăng, và từ 2011 đến 2020 dữ liệu cho thấy sự ổn định tương đối mặc dù vẫn có các điểm số ngoại lai. ###3.2.6.Biểu đồ Pareto

```r
library(dplyr)
library(ggplot2)

# Tính số lượng huy chương mỗi trường
huy_chuong_truong <- data %>%
  group_by(Hoc_sinh_truong) %>%
  summarise(So_huy_chuong = n()) %>%
  arrange(desc(So_huy_chuong)) %>%
  mutate(
    Truong = factor(Hoc_sinh_truong, levels = Hoc_sinh_truong),
    Cum_sum = cumsum(So_huy_chuong),
    Percent = Cum_sum / sum(So_huy_chuong) * 100
  )

# Vẽ biểu đồ Pareto
ggplot(huy_chuong_truong, aes(x = reorder(Truong, -So_huy_chuong))) +
  geom_col(aes(y = So_huy_chuong), fill = "steelblue") +
  geom_line(aes(y = Percent * max(So_huy_chuong) / 100, group = 1), color = "red", size = 1) +
  geom_point(aes(y = Percent * max(So_huy_chuong) / 100), color = "red", size = 2) +
  scale_y_continuous(
    name = "Số huy chương",
    sec.axis = sec_axis(~ . * 100 / max(huy_chuong_truong$So_huy_chuong), name = "Phần trăm tích lũy")
  ) +
  labs(
    title = "Biểu đồ Pareto - Các trường có nhiều huy chương nhất",
    x = "Trường"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

## 4.3 Phân nhóm các năm theo thành tích( cao - trung bình - thấp)


**Mục đích**: phân nhóm các năm theo thành tích (cao – trung bình – thấp) bằng thuật toán K-Means clustering.

**a. Gán trọng số cho từng loại huy chương**

Để phân nhóm công bằng hơn, ta quy đổi huy chương về một thước đo chung dựa trên trọng số:
- HCV:	4
- HCB:	3
- HCĐ:	2
- Bằng khen:	1

```r
# Tạo điểm thành tích dựa trên trọng số
data_nam <- df %>%
  group_by(Nam) %>%
  summarise(HCV = sum(Giai_thuong == "HCV"),
            HCB = sum(Giai_thuong == "HCB"),
            HCĐ = sum(Giai_thuong == "HCĐ"),
            BangKhen = sum(Giai_thuong == "Bằng khen"),
            Diem_tb = mean(Diem_so, na.rm = TRUE),
            Hang_tb = mean(Hang, na.rm = TRUE),
            Diem_thanh_tich = 4*HCV + 3*HCB + 2*HCĐ + 1*BangKhen,  # Tính điểm tổng hợp
            .groups = "drop")

# Xem dữ liệu đã tổng hợp
print(head(data_nam))
```

**b. Phân nhóm các năm bằng thuật toán K_means clustering**

```r
# Chuẩn hóa dữ liệu trước khi phân nhóm
data_scaled <- scale(data_nam[, c("Diem_thanh_tich", "Diem_tb", "Hang_tb")])

# Dùng K-Means để phân nhóm
set.seed(123)
kmeans_result <- kmeans(data_scaled, centers = 3, nstart = 25)

# Gán nhãn phân nhóm vào bảng dữ liệu gốc
data_nam$Cluster <- as.factor(kmeans_result$cluster)

# Xem kết quả
print(data_nam)
```

**c.Trực quan hóa lại kết quả phân nhóm**

```r
library(ggplot2)

ggplot(data_nam, aes(x = Nam, y = Diem_thanh_tich, color = Cluster)) +
  geom_point(size = 5, alpha = 0.7) +  # Hiển thị các điểm dữ liệu
  geom_smooth(method = "loess", se = FALSE, color = "black", linetype = "dashed") +  # Đường xu hướng
  theme_minimal() +
  labs(title = "Phân nhóm các năm theo thành tích tại IOI",
       x = "Năm",
       y = "Điểm thành tích",
       color = "Nhóm thành tích") +
  scale_color_manual(values = c("red", "blue", "green"))  # Màu sắc cho các nhóm
```

Nhận xét về biểu đồ phân nhóm các năm theo thành tích IOI:

1. Ba nhóm thành tích rõ rệt

- Nhóm 1 (màu đỏ): Điểm thành tích cao nhất, tập trung nhiều ở các năm sau 2000 và tăng mạnh giai đoạn gần đây.

- Nhóm 2 (màu xanh lá): Mức trung bình, phân bổ rải rác xen kẽ giữa Nhóm 1 và Nhóm 3.

- Nhóm 3 (màu xanh dương): Điểm thành tích thấp, thường xuất hiện ở giai đoạn đầu (cuối 1980 – đầu 1990).

2. Xu hướng tổng thể đi lên

- Đường xu hướng (đường nét đứt) cho thấy điểm thành tích của Việt Nam tại IOI tăng dần qua thời gian.

- Từ khoảng sau năm 2000 trở đi, đa số các điểm (năm) đều nằm ở mức cao (Nhóm 1), phản ánh sự cải thiện đáng kể.

- Giai đoạn thành tích thấp chủ yếu ở cuối thập niên 1980 – đầu 1990 (các chấm xanh dương). Sau đó, thành tích nâng lên rõ rệt, nhiều năm đạt mức cao (chấm đỏ) hơn.

3. Khoảng cách giữa các nhóm:

- Nhóm 3 có giá trị thấp hơn hẳn.

- Nhóm 1 có giá trị cao nhất và gia tăng về sau.

- Nhóm 2 nằm ở mức trung bình.

Kết luận:

- Việt Nam đã tiến bộ về thành tích IOI qua các năm, đặc biệt sau năm 2000.

- Giai đoạn gần đây (sau 2010) hầu như nằm ở nhóm thành tích cao, cho thấy sự đầu tư và đào tạo hiệu quả hơn.

## 4.4 Phân nhóm thành tích các trường

4.4.1 Tổng hợp dữ liệu theo trường

Trước tiên, ta sẽ nhóm dữ liệu theo trường (cột Hoc_sinh_truong) và tính các chỉ số chính:

- HCV, HCB, HCĐ, Bằng khen: Số lượng từng loại giải.

- Điểm thành tích (Weighted Score): Tính theo trọng số (ví dụ: HCV=4, HCB=3, HCĐ=2, Bằng khen=1).

- điểm trung bình (Diem_so) và xếp hạng trung bình (Hang)

```r
library(dplyr)

data_school <- df %>%
  group_by(Hoc_sinh_truong) %>%
  summarise(
    HCV = sum(Giai_thuong == "HCV"),
    HCB = sum(Giai_thuong == "HCB"),
    HCĐ = sum(Giai_thuong == "HCĐ"),
    BangKhen = sum(Giai_thuong == "Bằng khen"),
    Total_Awards = HCV + HCB + HCĐ + BangKhen,
    Weighted_Score = 4 * HCV + 3 * HCB + 2 * HCĐ + 1 * BangKhen,
    Avg_Score = mean(Diem_so, na.rm = TRUE),
    Avg_Rank = mean(Hang, na.rm = TRUE),
    .groups = "drop"
  )

head(data_school)
```

4.4.2 Phân nhóm thành tích giữa các trường

- Chúng ta dùng thuật toán K-Means để chia các trường thành 3 nhóm (cao, trung bình và thấp)

- Ở đây ta dùng các chỉ số như Weighted_Score, Avg_Score và Avg_Rank.

```r
# Chuẩn hóa các biến cần thiết trước khi clustering
data_school_scaled <- scale(data_school[, c("Weighted_Score", "Avg_Score", "Avg_Rank")])

set.seed(123)
kmeans_school <- kmeans(data_school_scaled, centers = 3, nstart = 25)

# Gán nhãn phân nhóm vào data
data_school$Cluster <- as.factor(kmeans_school$cluster)

head(data_school)
```

Giải thích:

- Sau đó, ta có thể xác định nhóm nào tương ứng với thành tích cao, trung bình hay thấp bằng cách so sánh giá trị trung bình của từng nhóm
- ví dụ: nhóm có Weighted_Score trung bình cao và Avg_Rank thấp là nhóm thành tích cao

4.4.3 Trực quan hóa so sánh thành tích giữa các trường

 Biểu đồ cột so sánh Weighted Score theo trường

**Sử dụng mô hình hồi quy tuyến tính (Linear Regression)**

Để dự đoán điểm trung bình trong tương lai, dựa trên xu hướng lịch sử được mô hình hóa bằng hồi quy tuyến tính đơn giản theo thời gian (năm).

```r
options(repos = c(CRAN = "https://cloud.r-project.org"))
install.packages("forecast", dependencies = TRUE)
```

```r
library(tidyverse)
library(lubridate)
library(ggplot2)
library(forecast)
# Mô hình hồi quy tuyến tính dự đoán điểm trung bình
model_diem <- lm(Diem_tb ~ Nam, data = data_nam)

# Dự đoán cho các năm tương lai (2025–2028)
nam_tuong_lai <- data.frame(Nam = 2025:2028)
du_doan_diem <- predict(model_diem, newdata = nam_tuong_lai)

# Tạo dataframe chứa kết quả dự đoán
du_doan_diem_df <- nam_tuong_lai %>%
  mutate(Diem_tb = du_doan_diem,
         Du_doan = TRUE)
# Gắn cờ cho dữ liệu gốc
data_nam_diem <- data_nam %>%
  select(Nam, Diem_tb) %>%
  mutate(Du_doan = FALSE)

# Kết hợp dữ liệu thực tế và dự đoán
data_diem_full <- bind_rows(data_nam_diem, du_doan_diem_df)
print(du_doan_diem_df[, c("Nam", "Diem_tb")])
```

```r
ggplot(data_diem_full, aes(x = Nam, y = Diem_tb)) +
  geom_line(size = 1.2) +
  geom_point(aes(color = Du_doan), size = 2) +
  scale_color_manual(values = c("FALSE" = "blue", "TRUE" = "red"),
                     labels = c("Thực tế", "Dự đoán")) +
  labs(
    title = "Xu hướng điểm trung bình của học sinh Việt Nam tại IOI",
    x = "Năm",
    y = "Điểm trung bình",
    color = "Dữ liệu"
  ) +
  theme_minimal()
```

```r
# Tải thư viện cần thiết
library(tidyverse)

# Chia dữ liệu thành tập huấn luyện (train) và tập kiểm tra (test)
train_data <- data_nam[1:(nrow(data_nam) - 4), ]  # 4 dữ liệu cuối làm test set
test_data <- data_nam[(nrow(data_nam) - 3):nrow(data_nam), ]  # 4 dữ liệu cuối làm test

# Mô hình hồi quy tuyến tính
model_rank <- lm(Hang_tb ~ Nam, data = train_data)

# Dự báo trên tập kiểm tra
du_doan_rank_test <- predict(model_rank, newdata = test_data)

# Tính các chỉ số độ chính xác
mae <- mean(abs(du_doan_rank_test - test_data$Hang_tb))  # Mean Absolute Error
rmse <- sqrt(mean((du_doan_rank_test - test_data$Hang_tb)^2))  # Root Mean Squared Error
mape <- mean(abs((du_doan_rank_test - test_data$Hang_tb) / test_data$Hang_tb)) * 100  # Mean Absolute Percentage Error

# Tính Accuracy (R-squared)
sse <- sum((test_data$Hang_tb - du_doan_rank_test)^2)  # Sum of Squared Errors
sst <- sum((test_data$Hang_tb - mean(test_data$Hang_tb))^2)  # Total Sum of Squares
accuracy <- 1 - (sse / sst)  # R-squared (Accuracy)

# In các chỉ số đánh giá
print(paste("MAE:", mae))
print(paste("RMSE:", rmse))
print(paste("MAPE:", mape))
print(paste("Accuracy (R-squared):", accuracy))

# Tạo khung năm tương lai để dự đoán
nam_tuong_lai <- data.frame(Nam = 2025:2028)
du_doan_rank <- predict(model_rank, newdata = nam_tuong_lai)

# Gộp vào khung dữ liệu
du_doan_rank_df <- nam_tuong_lai %>%
  mutate(Hang_tb = du_doan_rank,
         Du_doan = TRUE)

# Gắn cờ dữ liệu thật
data_nam_rank <- data_nam %>%
  select(Nam, Hang_tb) %>%
  mutate(Du_doan = FALSE)

# Kết hợp dữ liệu thật và dự đoán
data_rank_full <- bind_rows(data_nam_rank, du_doan_rank_df)

# In kết quả dự đoán cho các năm
print(du_doan_rank_df[, c("Nam", "Hang_tb")])

# Vẽ đồ thị kết quả dự đoán
ggplot(data_rank_full, aes(x = Nam, y = Hang_tb, color = Du_doan)) +
  geom_line() +
  geom_point() +
  ggtitle("Dự đoán hạng trung bình của học sinh Việt Nam tại IOI") +
  xlab("Năm") +
  ylab("Hạng trung bình (càng nhỏ càng tốt)") +
  theme_minimal()
```
